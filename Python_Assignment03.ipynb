```python
# ==========================================
# Task: Match Statement & Nested If/Else
# Contributed by: Aimen Ejaz Kiani
# ==========================================

# 1. MATCH STATEMENT EXAMPLES
print("--- Match Statement Examples ---")

# Example 1: Command handling
command = "save"
match command:
    case "open":
            print("Opening file...")
                case "save":
                        print("Saving file...")
                            case "exit":
                                    print("Exiting program...")
                                        case _:
                                                print("Unknown command")

                                                # Example 2: Month name
                                                month = 2
                                                match month:
                                                    case 1: print("January")
                                                        case 2: print("February")
                                                            case 3: print("March")
                                                                case _: print("Other month")

                                                                # Example 3: Fruit color
                                                                fruit = "Apple"
                                                                match fruit:
                                                                    case "Apple": print("Red")
                                                                        case "Banana": print("Yellow")
                                                                            case _: print("Color unknown")

                                                                            print("\n--- Difference: Match vs If/Else ---")
                                                                            print("1. Match is for pattern matching fixed values.")
                                                                            print("2. If/Else is better for ranges (e.g., age > 18).")

                                                                            # 2. NESTED IF/ELSE (5 EXAMPLES)
                                                                            print("\n--- Nested If/Else Examples ---")

                                                                            # Ex 1: Grading System
                                                                            marks = 85
                                                                            if marks >= 50:
                                                                                if marks >= 80:
                                                                                        print("Grade: A")
                                                                                            else:
                                                                                                    print("Grade: B")
                                                                                                    else:
                                                                                                        print("Fail")

                                                                                                        # Ex 2: Power Status
                                                                                                        power_on = True
                                                                                                        battery_full = False
                                                                                                        if power_on:
                                                                                                            if battery_full:
                                                                                                                    print("Device is ready and charged.")
                                                                                                                        else:
                                                                                                                                print("Device is on but battery is low.")
                                                                                                                                else:
                                                                                                                                    print("Device is off.")

                                                                                                                                    # Ex 3: Ticket Booking
                                                                                                                                    age = 25
                                                                                                                                    has_ticket = True
                                                                                                                                    if has_ticket:
                                                                                                                                        if age < 12:
                                                                                                                                                print("Child Entry Allowed")
                                                                                                                                                    else:
                                                                                                                                                            print("Adult Entry Allowed")
                                                                                                                                                            else:
                                                                                                                                                                print("Please buy a ticket first.")

                                                                                                                                                                # Ex 4: Internet Connection
                                                                                                                                                                wifi = True
                                                                                                                                                                connected = False
                                                                                                                                                                if wifi:
                                                                                                                                                                    if connected:
                                                                                                                                                                            print("Internet is working.")
                                                                                                                                                                                else:
                                                                                                                                                                                        print("Connected to WiFi, no internet.")
                                                                                                                                                                                        else:
                                                                                                                                                                                            print("WiFi is off.")

                                                                                                                                                                                            # Ex 5: Login Verification
                                                                                                                                                                                            is_registered = True
                                                                                                                                                                                            is_logged_in = True
                                                                                                                                                                                            if is_registered:
                                                                                                                                                                                                if is_logged_in:
                                                                                                                                                                                                        print("Access Granted!")
                                                                                                                                                                                                            else:
                                                                                                                                                                                                                    print("Please log in.")
                                                                                                                                                                                                                    else:
                                                                                                                                                                                                                        print("Please register first.")
```
