```python
# ðŸ“˜ PYTHON ASSIGNMENT 2: OPERATORS
# Name: Aimen Ejaz Kiani

# --- ARITHMETIC OPERATORS ---

# Q1. Write a Python program to add two numbers.
num1 = 10
num2 = 20
result = num1 + num2
print("Addition Result:", result)

# Q2. Find the remainder when one number is divided by another.
val1 = 15
val2 = 4
remainder = val1 % val2
print("Remainder:", remainder)

# Q3. Calculate the square of a number using the power operator (**).
base = 5
square = base ** 2
print("Square of 5 is:", square)


# --- ASSIGNMENT OPERATORS ---

# Q4. Assign a value and increase it by 5 using (+=).
x = 10
x += 5
print("Increased value:", x)

# Q5. Multiply a variable by 2 using (*=).
y = 8
y *= 2
print("Multiplied value:", y)

# Q6. Subtract 10 from a variable using (-=).
z = 50
z -= 10
print("Subtracted value:", z)


# --- COMPARISON OPERATORS ---

# Q7. Check whether two numbers are equal (==).
a = 15
b = 15
print("Are numbers equal?", a == b)

# Q8. Check if one number is greater than another (>).
print("Is 20 greater than 10?", 20 > 10)

# Q9. Check if first number is less than or equal to the second (<=).
num_a = 5
num_b = 10
print("Is 5 <= 10?", num_a <= num_b)


# --- LOGICAL OPERATORS ---

# Q10. Check if a number is greater than 5 AND less than 20.
num = 12
print("Is 12 between 5 and 20?", num > 5 and num < 20)

# Q11. Check if a number is less than 10 OR greater than 50.
val = 60
print("Is 60 < 10 or > 50?", val < 10 or val > 50)

# Q12. Use 'not' operator to reverse a condition.
is_raining = False
print("Is it NOT raining?", not is_raining)


# --- MEMBERSHIP OPERATORS ---

# Q13. Check if a number exists in a given list.
my_list = [1, 2, 3, 4, 5]
print("Is 3 in the list?", 3 in my_list)

# Q14. Check if a word exists in a given string.
sentence = "Python is amazing"
print("Does 'amazing' exist in string?", "amazing" in sentence)

# Q15. Check whether a value is NOT present in a tuple.
my_tuple = ("Apple", "Banana", "Cherry")
print("Is 'Orange' NOT in tuple?", "Orange" not in my_tuple)
```
